The dataset contains passenger details such as age, fare, class, and survival status.
There are missing values in Age, Cabin, and Embarked which require cleaning.

```python
# Convert Pclass to category
df['Pclass'] = df['Pclass'].astype('category')

# Cabin category (first letter)
df['Cabin_Category'] = df['Cabin'].str[0]

df.head()
```

```python
# Fill missing Age with median
df['Age'].fillna(df['Age'].median(), inplace=True)

df['Age'].isnull().sum()
```

```python
df['Family_Size'] = df['SibSp'] + df['Parch'] + 1

df.head()
```

Missing Age values were replaced with the median to reduce bias.
Family size gives better insight into group travel patterns.
Cabin categories help simplify complex cabin data.

```python
# Passenger with highest fare
df.loc[df['Fare'].idxmax()]
```

```python
# Passenger with largest family
df.loc[df['Family_Size'].idxmax()]
```

Passengers paying higher fares were typically in higher classes.
Large families may have influenced survival probability.

5. Family size indicates that moderate-sized families had better survival outcomes.
6. Missing data handling improved dataset consistency for analysis.
